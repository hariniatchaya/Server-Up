```python
features = [
    "Billing POS",
    "Inventory",
    "Online Ordering",
    "Digital Menu",
    "Analytics",
    "Loyalty Program",
    "Multi Branch",
    "Payment Integration"
]
feature_count = []
for feature in features:
    feature_count.append(
        df[feature].eq("Yes").sum()
    )
feature_df = pd.DataFrame({
    "Feature":features,
    "Companies":feature_count
})
print(feature_df)
plt.figure(figsize=(10,4))
plt.bar(feature_df["Feature"],
        feature_df["Companies"])
plt.xticks(rotation=45)
plt.ylabel("Companies")
plt.title("Feature Availability")
plt.show()
```

```python
features = [
    "Billing POS",
    "Inventory",
    "Online Ordering",
    "Digital Menu",
    "Analytics",
    "Loyalty Program",
    "Multi Branch",
    "Payment Integration"
]
```

```python
feature_df = df[features].replace({"YES":1, "NO":0})
df["Gap Score"] = len(features) - feature_df.sum(axis=1)
print(df[["Company","Gap Score"]])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Gap Score"], bins=8)
plt.title("Gap Score Distribution")
plt.xlabel("Gap Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
df["Strength Count"] = ( df["Strengths"].fillna("").str.split(",").str.len())
df["Weakness Count"] = (df["Weaknesses"].fillna("").str.split(",").str.len())
df["Review Score"] = (df["Strength Count"]-df["Weakness Count"])
print(df[["Company","Strength Count","Weakness Count","Review Score"]])
```
